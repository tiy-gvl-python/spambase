```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.cross_validation import train_test_split 
from sklearn.naive_bayes import MultinomialNB
```

```python
with open("spambase/spambase.names") as doc:
    vars = doc.readlines()[33:]
vars = [var.split(":")[0] for var in vars]
vars.append('spam_bool')
vars
```

```python
df = pd.read_csv("spambase/spambase.data", names=vars)
df.head()
```

## Using Sklearn train, test, split to Split matrices into test and train subsets.

```python
train, test = train_test_split(df, test_size=0.4)
```

```python
train.head()
```

```python
test.head()
```

```python
train_y = train[train.columns[-1]]
```

```python
train_y.head()
```

```python
test_y = test[test.columns[-1]]
```

```python
test_y.head()
```

```python
classifier = MultinomialNB()
```

```python
classifier.fit(train, train_y)
```

##Conclusion

###Given that our classifier score was greater than 0.8 we can determine that we can accurately predict the occurance of spam text.

```python
classifier.score(test, test_y)
```
